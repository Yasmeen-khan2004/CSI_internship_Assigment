```python
# # DATA CLEANING & PREPROCESSING
# ✅ Mount Google Drive and Load Dataset
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# ✅ Update this path based on your actual folder location in Drive
file_path = '/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv'

# Load the dataset
df = pd.read_csv(file_path)

# Preview
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# Define a function to create multiple plots for visualization
def visualize_titanic_data(data):
    plt.figure(figsize=(20, 25))

    # 1. Survival Count
    plt.subplot(3, 2, 1)
    sns.countplot(data=data, x='Survived', palette='Set2')
    plt.title('Survival Count (0 = No, 1 = Yes)')

    # 2. Survival by Gender
    plt.subplot(3, 2, 2)
    sns.countplot(data=data, x='Sex', hue='Survived', palette='Set1')
    plt.title('Survival by Gender')

    # 3. Age Distribution
    plt.subplot(3, 2, 3)
    sns.histplot(data=data, x='Age', bins=30, kde=True, color='skyblue')
    plt.title('Age Distribution')

    # 4. Pclass Distribution
    plt.subplot(3, 2, 4)
    sns.countplot(data=data, x='Pclass', hue='Survived', palette='muted')
    plt.title('Survival by Passenger Class')

    # 5. Fare Distribution
    plt.subplot(3, 2, 5)
    sns.histplot(data=data, x='Fare', bins=30, kde=True, color='orange')
    plt.title('Fare Distribution')

    # 6. Embarked vs Survival
    plt.subplot(3, 2, 6)
    sns.countplot(data=data, x='Embarked', hue='Survived', palette='coolwarm')
    plt.title('Survival by Embarkation Point')

    plt.tight_layout()
    plt.show()

# Call the visualization function
visualize_titanic_data(df)
```
